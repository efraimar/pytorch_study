```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
w = np.linspace(-10, 10)

loss = w**2 - 7*w -2

plt.plot(w, loss)
plt.xlabel('w')
plt.ylabel('loss')
plt.title('Loss Function')
plt.show()
```

```python
w = torch.tensor(1.0,requires_grad= True)
lerning_rate = 0.01
print(w)
for epoch in range(1000):
    loss = w**2 - 7*w -2
    loss.backward()
    if epoch % 10 == 0:
      print(f" w.grad = loss'(w): {w.grad} \t w: {w}")
    with torch.no_grad():
        w -= lerning_rate * w.grad
    w.grad.zero_()

print(w)
w = w.detach()
w = w.numpy()
x = np.linspace(-10, 10)
y = x**2 - 7*x -2
plt.plot(x,y)
y = w**2 - 7*w -2
plt.plot(w,y,'ro')
plt.xlabel('w')
plt.ylabel('loss')
plt.title('Loss Function')
plt.show()

```
